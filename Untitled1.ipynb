```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image_color = cv2.imread('img.jpg')

if image_color is not None:
    B, G, R = cv2.split(image_color)

    R_subtracted = cv2.subtract(R, 100)

    image_red_darker_BGR = cv2.merge([B, G, R_subtracted])

    image_original_RGB = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
    image_red_darker_RGB = cv2.cvtColor(image_red_darker_BGR, cv2.COLOR_BGR2RGB)

   
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image_original_RGB)
    plt.title(' ')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(image_red_darker_RGB)
    plt.title('   (-100)')
    plt.axis('off')

    plt.show()
else:
    print(":     'img.jpg'.")
```
