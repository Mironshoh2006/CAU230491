```

import numpy as np
import matplotlib.pyplot as plt

def rgb_to_cmy(image):
    return 1 - image

def cmy_to_rgb(image):
    return 1 - image

img = np.clip(np.random.rand(200, 300, 3), 0, 1)
cmy = rgb_to_cmy(img)
rgb_back = cmy_to_rgb(cmy)

plt.imshow(img)
plt.title("Original")
plt.show()

plt.imshow(cmy)
plt.title("CMY")
plt.show()
```

```

def intensity_slicing(gray):
    out = np.zeros((*gray.shape, 3))

    out[(gray < 0.2)] = [0, 0, 1]
    out[(gray >= 0.2) & (gray < 0.4)] = [0, 1, 1]
    out[(gray >= 0.4) & (gray < 0.6)] = [0, 1, 0]
    out[(gray >= 0.6) & (gray < 0.8)] = [1, 1, 0]
    out[(gray >= 0.8)] = [1, 0, 0]

    return out

gray = img.mean(axis=2)
sliced = intensity_slicing(gray)

plt.imshow(sliced)
plt.title("Intensity Slicing")
plt.show()
```

```

import cv2
from scipy.ndimage import gaussian_filter

rgb_smooth = gaussian_filter(img, sigma=2)

hsv = cv2.cvtColor((img*255).astype('uint8'), cv2.COLOR_RGB2HSV)
hsv[...,2] = gaussian_filter(hsv[...,2], sigma=2)
hsv_smooth = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

plt.imshow(rgb_smooth)
plt.title("RGB Smooth")
plt.show()

plt.imshow(hsv_smooth)
plt.title("HSV Smooth")
plt.show()
```
