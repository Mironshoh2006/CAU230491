```python
import cv2
import matplotlib.pyplot as plt


image = cv2.imread('img.jpg')

if image is None:
    print("Error: Could not find 'img.jpg'. Check the file path.")
else:
    
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

   
    subtracted_image = cv2.subtract(image_RGB, 100)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(image_RGB)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(subtracted_image)
    plt.title('Subtracted Image (-100)')
    plt.axis('off')

    plt.show()

```
