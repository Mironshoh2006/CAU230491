```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('img.jpg')
(h, w) = image.shape[:2]

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Positive translation (right, down)
T1 = np.float32([[1, 0, 100], [0, 1, 50]])
translated1 = cv2.warpAffine(image, T1, (w, h))

# Negative translation (left, up)
T2 = np.float32([[1, 0, -80], [0, 1, -40]])
translated2 = cv2.warpAffine(image, T2, (w, h))

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(translated1, cv2.COLOR_BGR2RGB))
plt.title("Translated (+100, +50)")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(translated2, cv2.COLOR_BGR2RGB))
plt.title("Translated (-80, -40)")
plt.axis('off')
plt.show()
```

```python
center = (w // 2, h // 2)

# Rotate 90 degrees
R1 = cv2.getRotationMatrix2D(center, 90, 1.0)
rot90 = cv2.warpAffine(image, R1, (w, h))

# Rotate 180 degrees
R2 = cv2.getRotationMatrix2D(center, 180, 1.0)
rot180 = cv2.warpAffine(image, R2, (w, h))

# Rotate with scaling
R3 = cv2.getRotationMatrix2D(center, 45, 0.7)
rot_scaled = cv2.warpAffine(image, R3, (w, h))

plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(rot90, cv2.COLOR_BGR2RGB))
plt.title("90 Rotation")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(cv2.cvtColor(rot180, cv2.COLOR_BGR2RGB))
plt.title("180 Rotation")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(rot_scaled, cv2.COLOR_BGR2RGB))
plt.title("45 + Scaling")
plt.axis('off')
plt.show()
```

```python
# Scaling factors
factors = [0.5, 1.5, 5]

for f in factors:
    S = cv2.getRotationMatrix2D((0, 0), 0, f)
    scaled = cv2.warpAffine(image, S, (int(w*f), int(h*f)))

    plt.imshow(cv2.cvtColor(scaled, cv2.COLOR_BGR2RGB))
    plt.title(f"Scaling Factor = {f}")
    plt.axis('off')
    plt.show()

    # Save large scaling result
    if f >= 5:
        cv2.imwrite(f"scaled_{f}.png", scaled)
        
```

```python
# X-axis shearing
mx = 0.6
shear_x = np.float32([[1, mx, 0], [0, 1, 0]])
sheared_x = cv2.warpAffine(image, shear_x, (w, h))

# Y-axis shearing
my = 0.5
shear_y = np.float32([[1, 0, 0], [my, 1, 0]])
sheared_y = cv2.warpAffine(image, shear_y, (w, h))

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(sheared_x, cv2.COLOR_BGR2RGB))
plt.title("Shearing X-axis")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(sheared_y, cv2.COLOR_BGR2RGB))
plt.title("Shearing Y-axis")
plt.axis('off')
plt.show()
```
